```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import matplotlib.pylab as plt
import pandas as pd, numpy as np
from sklearn.metrics import mean_absolute_error
from hyperopt import hp, tpe, STATUS_OK, Trials,fmin
import hyperopt
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd
d=pd.read_excel('data2.xlsx')
k=d[['x1','x2']]
l=d.y
a=0
b=0
p=0
q=0

r=p*(d.x1**a)+q*(d.x2**b)

def f(space):
    a = space['a']
    b= space['b']
    p= space['p']
    q= space['q']
    loss = mean_absolute_error(l,r)
    result = {
        "loss": loss,
        "status": STATUS_OK,
        "space": space
    }
    return result

space = {
    'a': hp.quniform('a', 0, 0.1,0.01),
    'b': hp.quniform('b', -1, -0.5,0.01),
    'p': hp.quniform('p', 0, 5,0.01),
    'q': hp.quniform('q', 0, 10,0.01),
    
    
}

trials = Trials()

best = fmin(
    fn=f,
    space=space,
    algo=tpe.suggest,
    trials=trials,
    max_evals=1000
)
print(best)

g=(d.x1)**0.0005
h=(d.x2)**-0.56
j=pd.DataFrame({"g":g,"h":h})
results = sm.OLS(l,j).fit()
print(results.summary())

```
