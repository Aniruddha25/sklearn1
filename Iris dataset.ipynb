```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.feature_extraction import DictVectorizer

from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import f_regression,chi2
iris=load_iris()
iris.keys()
d=pd.DataFrame(iris.data,columns=iris.feature_names)
X,y= iris.data,iris.target

pd.plotting.scatter_matrix(d,c=iris.target,figsize=(10,10));
classifier=KNeighborsClassifier()
train_X,test_X,train_y,test_y= train_test_split(X,y,train_size=0.6,test_size=0.4,stratify=y)
print("training")
print(train_y)
print("testing")
print(test_y)
classifier.fit(train_X,train_y)
pred_y=classifier.predict(test_X)
print('Accuracy')
print("  ")
print(classifier.score(test_X,test_y))
print("SAMPLE CLASSIFICATION :")
print('      ')
print("1.Samples correcrly classified:")
correct_idx=np.where(pred_y==test_y)[0]
print(correct_idx)
print('  ')
print("2.Samples incorrectly classified:")
incorrect_idx=np.where(pred_y!=test_y)[0]
print(incorrect_idx)




```

```python
print("FEATURE SELECTION-")
F=f_regression(X,y)
print('1.F-statistics')
print(F)
print("   ")
print("2.chi disribution")
chi=chi2(X,y)
print(chi)
```
